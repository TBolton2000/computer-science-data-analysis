# What is Computer Science?
Our team of three computer science majors decided to collect some data about computer science to learn about what we need to know, what different topics are prominent, and what we can do to get ahead.

But first we needed to collect data of our own.

From previous experience, our team has used coding practice problem websites to prepare for technical interviews. 

We decided to collect a large amounts of words from these problems to analyze their frequency.

```python
f = open("leet_data.txt", mode = 'r', encoding = "utf-8")
lines = f.readlines()
f.close()
formatted_lines = []
#Try to format all of the text to remove unnecessary characters
for line in lines:
    formatted_lines.append(line.strip().upper().replace('.','').replace(':','').replace('(','').replace(')','').replace(',',''))

formatted_lines
```

```python
#Put all of the words into a single list
all_words = []
for line in formatted_lines:
    all_words += line.split(' ')

all_words
```

```python
#Count all word frequencies
f = open("new_out.txt",mode = "w",encoding = "utf-8")
counts = {}
for word in all_words:
    f.write(word+" ")
f.close()
for word in all_words:
    if word in counts:
        counts[word]+=1
    else:
        counts[word]=1

counts
```

```python
print(sorted(counts.items(), key = lambda kv:(kv[1], kv[0]),reverse = True))
counts.pop('THE')
counts.pop('A')
counts.pop('OF')
counts.pop('')
counts.pop('AND')
counts.pop('IS')
counts.pop('TO')
counts.pop('IN')
counts.pop('GIVEN')
counts.pop('YOU')
counts.pop('AN')
counts.pop('THAT')
counts.pop('ALL')
counts.pop('ARE')
counts
```
